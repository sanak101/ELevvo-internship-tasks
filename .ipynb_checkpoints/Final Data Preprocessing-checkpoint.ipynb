```python
from Bio import SeqIO

# Function to count sequences in a FASTA file
def count_sequences(filepath):
    return sum(1 for _ in SeqIO.parse(filepath, "fasta"))

# File paths (replace with the actual paths of your files)
autophagy_train_path = 'autophagy.trn.fasta'
autophagy_test_path = 'autophagy.tst.fasta'
non_autophagy_train_path = 'non-autophagy.trn.fasta'
non_autophagy_test_path = 'non-autophagy.tst.fasta'

# Count sequences in each dataset
autophagy_train_count = count_sequences(autophagy_train_path)
autophagy_test_count = count_sequences(autophagy_test_path)
non_autophagy_train_count = count_sequences(non_autophagy_train_path)
non_autophagy_test_count = count_sequences(non_autophagy_test_path)

# Calculate total number of sequences
total_sequences = (autophagy_train_count + autophagy_test_count +
                   non_autophagy_train_count + non_autophagy_test_count)

# Calculate percentage of total data for each set
autophagy_train_percentage = (autophagy_train_count / total_sequences) * 100
autophagy_test_percentage = (autophagy_test_count / total_sequences) * 100
non_autophagy_train_percentage = (non_autophagy_train_count / total_sequences) * 100
non_autophagy_test_percentage = (non_autophagy_test_count / total_sequences) * 100

# Print the results
print(f"Autophagy Train: {autophagy_train_count} sequences ({autophagy_train_percentage:.2f}% of total data)")
print(f"Autophagy Test: {autophagy_test_count} sequences ({autophagy_test_percentage:.2f}% of total data)")
print(f"Non-Autophagy Train: {non_autophagy_train_count} sequences ({non_autophagy_train_percentage:.2f}% of total data)")
print(f"Non-Autophagy Test: {non_autophagy_test_count} sequences ({non_autophagy_test_percentage:.2f}% of total data)")

# Print total sequences for reference
print(f"Total Sequences: {total_sequences}")

#OUR NON AUTOPHAGY DATA IS MORE THAN AUTOPHAGY SO WE WILL APPLY TECHNIQUES LIKE SMOTE OR UNDERSAMPLING TO ELIMINATE CLASS
#IMBALANCES
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from Bio import SeqIO

# Function to load FASTA and return a DataFrame
def load_fasta_to_dataframe(filepath, label_type):
    records = []
    for record in SeqIO.parse(filepath, "fasta"):
        records.append([str(record.id), str(record.seq), label_type])
    df = pd.DataFrame(records, columns=['ID', 'Sequence', 'Label'])
    return df

# Load datasets
autophagy_train_df = load_fasta_to_dataframe('autophagy.trn.fasta', 'autophagy_train')
autophagy_test_df = load_fasta_to_dataframe('autophagy.tst.fasta', 'autophagy_test')
non_autophagy_train_df = load_fasta_to_dataframe('non-autophagy.trn.fasta', 'non_autophagy_train')
non_autophagy_test_df = load_fasta_to_dataframe('non-autophagy.tst.fasta', 'non_autophagy_test')

# Calculate sequence character counts
autophagy_train_df['seq_char_count'] = autophagy_train_df['Sequence'].apply(len)
autophagy_test_df['seq_char_count'] = autophagy_test_df['Sequence'].apply(len)
non_autophagy_train_df['seq_char_count'] = non_autophagy_train_df['Sequence'].apply(len)
non_autophagy_test_df['seq_char_count'] = non_autophagy_test_df['Sequence'].apply(len)

# Function to plot sequence counts
def plot_seq_count(df, data_name):
    sns.histplot(df['seq_char_count'], kde=True, bins=30)
    plt.title(f'Sequence Length Distribution: {data_name}')
    plt.xlabel('Length of Sequence')
    plt.ylabel('Frequency')
    plt.grid(True)

# Create subplots for all datasets
plt.figure(figsize=(18, 5))

plt.subplot(1, 4, 1)
plot_seq_count(autophagy_train_df, 'Autophagy Train')

plt.subplot(1, 4, 2)
plot_seq_count(autophagy_test_df, 'Autophagy Test')

plt.subplot(1, 4, 3)
plot_seq_count(non_autophagy_train_df, 'Non-Autophagy Train')

plt.subplot(1, 4, 4)
plot_seq_count(non_autophagy_test_df, 'Non-Autophagy Test')

plt.subplots_adjust(right=2.0)
plt.show()
```

```python
import numpy as np
from Bio import SeqIO
from imblearn.over_sampling import SMOTE
from scipy.sparse import csr_matrix, vstack

def extract_alphabet(fasta_files):
    """Extract unique characters (amino acids/nucleotides) from given FASTA files."""
    unique_chars = set()
    for file_path in fasta_files:
        for record in SeqIO.parse(file_path, "fasta"):
            unique_chars.update(str(record.seq))
    return "".join(sorted(unique_chars))

def one_hot_encode_sequence(sequence, alphabet, max_len):
    """One-hot encode a sequence and pad it to max_len."""
    char_to_int = {char: idx for idx, char in enumerate(alphabet)}
    encoding = np.zeros((max_len, len(alphabet)), dtype=int)
    for i, char in enumerate(sequence[:max_len]):  # Truncate if sequence exceeds max_len
        if char in char_to_int:
            encoding[i, char_to_int[char]] = 1
    return encoding.flatten()

def load_and_encode_fasta(file_path, label, alphabet, max_len, batch_size=32):
    """Load sequences from a FASTA file, pad them, and apply one-hot encoding in batches."""
    batch_sequences, batch_labels = [], []

    for record in SeqIO.parse(file_path, "fasta"):
        encoded_seq = one_hot_encode_sequence(str(record.seq), alphabet, max_len)
        batch_sequences.append(encoded_seq)
        batch_labels.append(label)

        # When batch size is reached, yield the batch
        if len(batch_sequences) == batch_size:
            X_batch = csr_matrix(np.array(batch_sequences))
            y_batch = np.array(batch_labels)
            yield X_batch, y_batch
            batch_sequences, batch_labels = [], []

    # Yield remaining data if not exactly divisible by batch size
    if batch_sequences:
        X_batch = csr_matrix(np.array(batch_sequences))
        y_batch = np.array(batch_labels)
        yield X_batch, y_batch

def get_max_sequence_length(fasta_files):
    """Find the maximum sequence length from given FASTA files."""
    max_len = 0
    for file_path in fasta_files:
        for record in SeqIO.parse(file_path, "fasta"):
            max_len = max(max_len, len(record.seq))
    return max_len

def main():
    # Paths to FASTA files
    autophagy_train_path = "autophagy.trn.fasta"
    non_autophagy_train_path = "non-autophagy.trn.fasta"

    # Extract the alphabet dynamically and find max sequence length
    fasta_files = [autophagy_train_path, non_autophagy_train_path]
    alphabet = extract_alphabet(fasta_files)
    max_len = get_max_sequence_length(fasta_files)
    print(f"Alphabet extracted from data: {alphabet}")
    print(f"Maximum sequence length: {max_len}")

    # Initialize lists to hold all data
    all_sequences, all_labels = [], []

    # Load and encode Autophagy data in batches
    print("Loading and encoding Autophagy data in batches...")
    for batch_X, batch_y in load_and_encode_fasta(autophagy_train_path, label=1, alphabet=alphabet, max_len=max_len, batch_size=32):
        all_sequences.append(batch_X)
        all_labels.append(batch_y)

    # Load and encode Non-Autophagy data in batches
    print("Loading and encoding Non-Autophagy data in batches...")
    for batch_X, batch_y in load_and_encode_fasta(non_autophagy_train_path, label=0, alphabet=alphabet, max_len=max_len, batch_size=32):
        all_sequences.append(batch_X)
        all_labels.append(batch_y)

    # Combine all batches
    X_train = vstack(all_sequences)  # Stack all sparse matrices vertically
    y_train = np.hstack(all_labels)  # Combine all labels into a single array

    # Print original class distribution
    print(f"Original class distribution: {np.bincount(y_train)}")

    # Apply SMOTE after combining all data
    smote = SMOTE(random_state=42)
    print("Applying SMOTE...")
    X_train_final, y_train_final = smote.fit_resample(X_train, y_train)

    # Print balanced class distribution
    print(f"Balanced class distribution: {np.bincount(y_train_final)}")

    # Print sparse matrix summary of balanced data
    print("\nSummary of balanced data (sparse matrix):")
    print(f"Number of samples: {X_train_final.shape[0]}")
    print(f"Number of features: {X_train_final.shape[1]}")
    print(f"Non-zero elements in the sparse matrix: {X_train_final.nnz}")

    # Print a few samples (non-zero elements) from the sparse matrix
    print("\nNon-zero elements in the first 5 samples:")
    for i in range(min(5, X_train_final.shape[0])):
        row = X_train_final.getrow(i)
        print(f"Sample {i + 1} - Label: {y_train_final[i]}")
        print(f"Non-zero elements (feature indices): {row.indices}")
        print("-" * 80)

    # Save the sparse matrix (X_train_final) and labels (y_train_final) to files
    print("\nSaving data to disk...")

    # Save sparse matrix as .npz (Compressed Sparse Matrix format)
    from scipy.sparse import save_npz
   # Save sparse matrix as .npz (Compressed Sparse Matrix format)
    save_npz('X_train_final.npz', X_train_final)

    # Save labels as .npy
    np.save('y_train_final.npy', y_train_final)

    print("Data saved successfully!")

    # Return the values for further use
    return X_train_final, y_train_final

# Call main() and store the returned values
X_train_final, y_train_final = main()

# Optionally, print the returned values
print("Returned X_train_final shape:", X_train_final.shape)
print("Returned y_train_final shape:", y_train_final.shape)
```
